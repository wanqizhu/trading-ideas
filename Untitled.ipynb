```python
start_wealth = 100
start_money = start_wealth / 2
start_coin = start_wealth / 2 / df_day_prices['Open'][0]
trade_fee_perct = 0.001
size_per_trade = 0.1  # fraction of total money to put in per bid/ask
cancel_limit = 1.2  # for buys, cancel if buy_price * cancel_limit < price
                    # for sells, cancel if sell_price / cancel_limit > price


# contains orders of the form (price, quantity)
open_buys = set()
open_sells = set()
wealth_over_time = []  # these arrays track start-of-time-window statistics
money_over_time = []
coin_wealth_over_time = []
pending_buys_money_over_time = []
pending_sells_coin_wealth_over_time = []

spread_perct = 0.05

money = start_money
coin = start_coin
pending_buys_money = 0
pending_sells_coin = 0

"""
Simulate the following strategy:

At open, we put out buy/sell orders with `spread_perct` spread, based on the current open price, if we have
liquitity. The trade size in dollar ammounts in `size_per_trade * total_wealth`.
At close, we take off any remaining orders to put them back in at next day's open.


To check if an order has executed during the day, we only have to look at that day's high/low.

"""

# TODO: round money to cents
for day, prices in df_day_prices.iterrows():
    print(f"Day {day}")
    # put back orders from previous days, checking if any will execute
    to_buy = [b for b in open_buys if b[0] > prices['Open']]
    for buy in to_buy:
        open_buys.remove(buy)
        coin += buy[1] * (1 / (1 + trade_fee_perct))
        pending_buys_money -= buy[0] * buy[1]
        print(f"Bought {buy[1]} at price {buy[0]} for total cost {buy[0] * buy[1]}")
    
    to_sell = [s for s in open_sells if s[0] < prices['Open']]
    for sell in to_sell:
        open_sells.remove(sell)
        money += sell[0] * sell[1] * (1 / (1 + trade_fee_perct))
        pending_sells_coin -= sell[1]
        print(f"Sold {sell[1]} at price {sell[0]} for total gain {sell[0] * sell[1]}")

        
        
    
    # cancel any orders that are way too out of line with price
    to_cancel = [b for b in open_buys if b[0] * cancel_limit < prices['Open']]
    for order in to_cancel:
        open_buys.remove(order)
        money += order[0] * order[1]
        pending_buys_money -= order[0] * order[1]
        
    to_cancel = [s for s in open_sells if s[0] / cancel_limit > prices['Open']]
    for order in to_cancel:
        open_sells.remove(order)
        coin += order[1]
        pending_sells_coin -= order[1]
    
    
    
    # add new bid/ask spread based on open price
    # first, figure out current total wealth and cash available at start of day
    total_wealth = money + pending_buys_money + (coin + pending_sells_coin) * prices['Open']
    wealth_over_time.append(total_wealth)
    money_over_time.append(money)
    coin_wealth_over_time.append(coin * prices['Open'])
    pending_buys_money_over_time.append(pending_buys_money)
    pending_sells_coin_wealth_over_time.append(pending_sells_coin * prices['Open'])
    
    trade_size = total_wealth * size_per_trade
    
    if trade_size < money:  # have money for a buy order
        buy_price = prices['Open'] * (1 - spread_perct/2)
        buy_quantity = trade_size / buy_price
        open_buys.add((buy_price, buy_quantity))
        money -= trade_size
        pending_buys_money += trade_size
    
    sell_price = prices['Open'] * (1 + spread_perct/2)
    sell_quantity = trade_size / sell_price
    if sell_quantity < coin:  # have coin for a sell order
        open_sells.add((sell_price, sell_quantity))
        coin -= sell_quantity
        pending_sells_coin += sell_quantity
        
    
    # end-of-day: check if today's min/max executed any orders
    to_buy = [b for b in open_buys if b[0] > prices['Low']]
    for buy in to_buy:
        open_buys.remove(buy)
        coin += buy[1] * (1 / (1 + trade_fee_perct))
        pending_buys_money -= buy[0] * buy[1]
        print(f"Bought {buy[1]} at price {buy[0]} for total cost {buy[0] * buy[1]}")

    
    to_sell = [s for s in open_sells if s[0] < prices['High']]
    for sell in to_sell:
        open_sells.remove(sell)
        money += sell[0] * sell[1] * (1 / (1 + trade_fee_perct))
        pending_sells_coin -= sell[1]
        print(f"Sold {sell[1]} at price {sell[0]} for total gain {sell[0] * sell[1]}")
```

```python
plt.plot(wealth_over_time, label='wealth')
plt.plot(money_over_time, label='liquid cash')
plt.plot(coin_wealth_over_time, label='liquid coin value')
plt.plot(pending_buys_money_over_time, label='pending buys value')
plt.plot(pending_sells_coin_wealth_over_time, label='pending sells value')
plt.plot(df_day_prices['Open'] / 100, label='coin price')
plt.legend()
```

```python
start_wealth = 100
start_money = start_wealth / 2
start_coin = start_wealth / 2 / df_day_prices['Open'][0]
trade_fee_perct = 0.001
size_per_trade = 0.25  # fraction of total money to put in per bid/ask
cancel_limit = 10  # for buys, cancel if buy_price * cancel_limit < price
                    # for sells, cancel if sell_price / cancel_limit > price


# contains orders of the form (price, quantity)
open_buys = set()
open_sells = set()
wealth_over_time = []  # these arrays track start-of-time-window statistics
money_over_time = []
coin_wealth_over_time = []
pending_buys_money_over_time = []
pending_sells_coin_wealth_over_time = []

spread_perct = 0.05

money = start_money
coin = start_coin
pending_buys_money = 0
pending_sells_coin = 0

"""
Simulate the following strategy:

At open, we put out buy/sell orders with `spread_perct` spread, based on the current open price, if we have
liquitity. The trade size in dollar ammounts in `size_per_trade * total_wealth`.
At close, we cancel the two orders we put out at open.

Whenever a buy order executes, we put out a permanent sell order at +spread%.
Whenever a sell order executes, we put out a permanent buy order at -spread%

To check if an order has executed during the day, we only have to look at that day's high/low.

"""


for day, prices in df_day_prices.iterrows():
    print(f"Day {day}: open {prices['Open']}, low {prices['Low']}, high {prices['High']}, close {prices['Close']}")
    
    # cancel any orders that are way too out of line with price
    to_cancel = [b for b in open_buys if b[0] * cancel_limit < prices['Open']]
    for order in to_cancel:
        open_buys.remove(order)
        money += order[0] * order[1]
        pending_buys_money -= order[0] * order[1]
        
    to_cancel = [s for s in open_sells if s[0] / cancel_limit > prices['Open']]
    for order in to_cancel:
        open_sells.remove(order)
        coin += order[1]
        pending_sells_coin -= order[1]
    
    
    # add new bid/ask spread based on open price
    # first, figure out current total wealth and cash available at start of day
    total_wealth = money + pending_buys_money + (coin + pending_sells_coin) * prices['Open']
    wealth_over_time.append(total_wealth)
    money_over_time.append(money)
    coin_wealth_over_time.append(coin * prices['Open'])
    pending_buys_money_over_time.append(pending_buys_money)
    pending_sells_coin_wealth_over_time.append(pending_sells_coin * prices['Open'])
    
    trade_size = total_wealth * size_per_trade
    
    today_buy_order = None
    today_sell_order = None
    
    if trade_size < money:  # have money for a buy order
        buy_price = prices['Open'] * (1 - spread_perct/2)
        buy_quantity = trade_size / buy_price
        today_buy_order = (buy_price, buy_quantity)
    
    sell_price = prices['Open'] * (1 + spread_perct/2)
    sell_quantity = trade_size / sell_price
    if sell_quantity < coin:  # have coin for a sell order
        today_sell_order = (sell_price, sell_quantity)

        
    # end-of-day: check if today's min/max executed any permanent orders
    to_buy = [b for b in open_buys if b[0] > prices['Low']]
    for buy in to_buy:
        open_buys.remove(buy)
        coin += buy[1] * (1 / (1 + trade_fee_perct))
        pending_buys_money -= buy[0] * buy[1]
        print(f"Bought {buy[1]} at price {buy[0]} for total cost {buy[0] * buy[1]}")
    
    to_sell = [s for s in open_sells if s[0] < prices['High']]
    for sell in to_sell:
        open_sells.remove(sell)
        money += sell[0] * sell[1] * (1 / (1 + trade_fee_perct))
        pending_sells_coin -= sell[1]
        print(f"Sold {sell[1]} at price {sell[0]} for total gain {sell[0] * sell[1]}")
        
    # check if today's min/max executed today's orders
    # note that these orders have not really been "paid", since if they don't execute here they get cancelled
    if today_buy_order and today_buy_order[0] > prices['Low']:
        coin_quantity = today_buy_order[1] * (1 / (1 + trade_fee_perct))
        money -= today_buy_order[0] * today_buy_order[1]
        print(f"Bought {today_buy_order[1]} at price {today_buy_order[0]} for total cost {today_buy_order[0] * today_buy_order[1]}")
        # add permanent sell order
        sell_price = today_buy_order[0] * (1 + spread_perct)
        open_sells.add((sell_price, coin_quantity))
        pending_sells_coin += coin_quantity
        
    if today_sell_order and today_sell_order[0] < prices['High']:
        coin -= today_sell_order[1]
        money_gained = today_sell_order[0] * today_sell_order[1] * (1 / (1 + trade_fee_perct))
        print(f"Sold {today_sell_order[1]} at price {today_sell_order[0]} for total cost {today_sell_order[0] * today_sell_order[1]}")
        
        # add permanent buy order
        buy_price = today_sell_order[0] * (1 - spread_perct)
        open_buys.add((buy_price, money_gained / buy_price))
        pending_buys_money += money_gained
```

```python
plt.plot(wealth_over_time, label='wealth')
plt.plot(money_over_time, label='liquid cash')
plt.plot(coin_wealth_over_time, label='liquid coin value')
plt.plot(pending_buys_money_over_time, label='pending buys value')
plt.plot(pending_sells_coin_wealth_over_time, label='pending sells value')
plt.legend()
```

```python
# constant rebalance strategy
# at end of each day, rebalance porfolio to have equal value in cash/coin
start_wealth = 100
start_money = start_wealth / 2
start_coin = start_wealth / 2 / df_day_prices['Open'][0]
trade_fee = 0.001

money = start_money
coin = start_coin


wealth_over_time = []  # these arrays track end-of-time-window statistics


for day, prices in df_day_prices.iterrows():
    # end-of-day: rebalance
    coin_value = coin * prices['Close']
    if coin_value > money:
        # sell some coin
        # b/c trading fees, this is not a perfect rebalance, but it's close enough
        coin_to_sell = (coin_value - money) / prices['Close'] / 2
        coin -= coin_to_sell
        money += coin_to_sell * prices['Close'] * (1 / (1 + trade_fee))
    
    elif coin_value < money:
        # buy some coin
        # b/c trading fees, this is not a perfect rebalance, but it's close enough
        coin_to_buy = (money - coin_value) / prices['Close'] / 2
        coin += coin_to_buy
        money -= coin_to_buy * prices['Close'] * (1 + trade_fee)
    
    # figure out current total wealth
    total_wealth = money + coin * prices['Close']
    wealth_over_time.append(total_wealth)
```

```python
plt.plot(wealth_over_time, label='wealth')
plt.plot(df_day_prices['Close'] / 100)
plt.legend()
```
